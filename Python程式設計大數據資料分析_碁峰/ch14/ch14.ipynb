```python
# %load Matplotlib__1.py
import matplotlib.pyplot as plt

# A
companyA = {'Personnel Department': 8,  # 
            'Accounting Department': 5,  # 
            'Design Department': 10,  # 
            'Marketing Department': 12,  # 
            'Purchasing Department': 5}  # 
# B
companyB = {'Personnel Department': 10,  # 
            'Accounting Department': 8,  # 
            'Design Department': 20,  # 
            'Marketing Department': 18,  # 
            'Purchasing Department': 12}  # 

# A
companyA_values = companyA.values()
# B
companyB_values = companyB.values()
# 
keys = companyA.keys()
# 
colors = ['violet', 'yellow', 'skyblue', 'lightcoral', 'lightgreen']
# ""
explode = [0, 0, 0, 0.1, 0]
# 
autopct = '%.1f%%'
# 
startangle = 90

# A
fig1, ax1 = plt.subplots()
ax1.pie(companyA_values,
        labels=keys,
        explode=explode,
        colors=colors,
        autopct=autopct,
        startangle=startangle)
# 
ax1.axis('equal')
# "Company A"20
ax1.set_title('Company A', size=20)

# B
fig2, ax2 = plt.subplots()
ax2.pie(companyB_values,
        labels=keys,
        explode=explode,
        colors=colors,
        autopct=autopct,
        startangle=startangle)
# 
ax2.axis('equal')
# "Company B"20
ax2.set_title('Company B', size=20)

# 
plt.show()
```

```python
# %load Matplotlib__1.py
from matplotlib import pyplot as plt
import numpy as np

# 
scores = [84, 83, 23, 63, 45, 43, 72, 65, 60, 73,
          34, 26, 59, 20, 31, 63, 45, 79, 98, 35,
          61, 76, 20, 90, 30, 45, 44, 92, 53, 93,
          67, 33, 38, 24, 45, 46, 39, 49, 56, 75,
          47, 72, 60, 40, 91, 69, 96, 49, 25, 35,
          64, 20, 43, 65, 72, 78, 28, 53, 31, 100,
          41, 65, 35, 51, 40, 37, 79, 69, 54, 49]

# range_count[0]: range0~19
# range_count[1]: range20~39
# range_count[2]: range40~59
# range_count[3]: range60~79
# range_count[4]: range80~100
# 0
range_count = [0] * 5

# 
for score in scores:
  if score < 20:
    range_count[0] += 1
  elif score < 40:
    range_count[1] += 1
  elif score < 60:
    range_count[2] += 1
  elif score < 80:
    range_count[3] += 1
  else:
    range_count[4] += 1

# y
index = np.arange(0, 25, 5)
# x
labels = ['0~19', '20~39', '40~59', '60~79', '80~100']
# 
plt.bar(index, range_count, width=2)
# x
plt.xlabel('Range', fontsize=14)
# y
plt.ylabel('Quantity', fontsize=14)
# x
plt.xticks(index, labels)
# y
plt.yticks(index)
# 
plt.title('Score ranges count', fontsize=20)
# 
plt.show()
```

```python
# %load Matplotlib__2.py
import csv
import matplotlib.pyplot as plt

city = []
population = []

# 
with open('data/csvfile3.csv') as file:
  data = list(csv.reader(file))
  for record in data[1:]:
    city.append(record[0])
    population.append(record[1])

# 
keys = city
# 
colors = ['violet', 'yellow', 'skyblue', 'lightcoral', 'lightgreen', 'gold']
# "New Taipei City"
explode = [0.1, 0, 0, 0.1, 0, 0]
# 
autopct = '%.1f%%'
# 
startangle = 90

# 
plt.pie(population,
        labels=keys,
        explode=explode,
        colors=colors,
        autopct=autopct,
        startangle=startangle)
# 
plt.axis('equal')
# 20
plt.title('Provincial Cities Population', size=20)
# 
plt.show()
```

```python
# %load Matplotlib__2.py
import numpy as np
import matplotlib.pyplot as plt

# 
male = [20, 48, 52, 31, 22]
# 
female = [19, 50, 44, 25, 20]

# 
index = np.arange(len(male))
width = 0.35

fig, ax = plt.subplots()
# 
rects1 = ax.bar(index, male, width=0.35, color='r')
# 
rects2 = ax.bar(index + width, female, width, color='y')

# xy
ax.set_title('Quantity by group and gender', fontsize=18)
ax.set_xlabel('Age range', fontsize=14)
ax.set_ylabel('Quantity', fontsize=14)
ax.set_xticks(index + width / 2)
ax.set_xticklabels(['10~15', '16~20', '21~25', '25~30', '31~35'])
ax.set_yticks(np.arange(0, 51, 10))

# 
ax.legend((rects1[0], rects2[0]), ['Male', 'Female'])

plt.show()
```

```python
# %load Matplotlib__2.py
import csv
import numpy as np
from matplotlib import pyplot as plt

x_labels = []  # x
# csvfile1.csv
with open('data/csvfile1.csv', 'r') as csvfile:
  # csv.reader
  rows = list(csv.reader(csvfile, delimiter=','))
  # 
  categories = rows[0][1:]
  # 
  categories_profit = [[0] * (len(rows)-1) for _ in range(len(categories))]
  # 
# x
# 
  # ij
  for i in range(1, len(rows)):
    x_labels.append(rows[i][0])
    for j in range(len(categories)):
      categories_profit[j][i-1] = eval(rows[i][j+1])

lines = []
x = np.arange(1, 5)
# '-o'()2
for i in range(len(categories)):
  # 
  line, = plt.plot(x, categories_profit[i], '-o', linewidth=2)
  lines.append(line)

# x141
plt.xticks(np.arange(1, 5, 1), x_labels)
# y025050
plt.yticks(np.arange(0, 251, 50))
# x16
plt.xlabel('Season', size=16)
# y16
plt.ylabel('Thousand Dollars', size=16)
# 16
plt.title('Profit earned in one year', size=20)
# 
plt.legend(lines, categories)
# 
plt.show()
```
